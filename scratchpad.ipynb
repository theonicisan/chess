```python

dictionary = np.load('dictionary.npy',allow_pickle='TRUE').item()
multiple_counter = 0
counter = 0
highest = 0
position = ''

for key, value in dictionary.items():
    if value[1] > 1:
        #print(key, ' : ', value)
        multiple_counter += 1
        print(key, ' : ', value)
        if value[1] > highest:
            highest = value[1]
            position = key

    counter += 1

#print(dictionary)
print('Number of positions in dictionary: ' + str(counter))
print('Number of positions that have occurred more than once: ' + str(multiple_counter))
print('Highest occurring position: ' + str(highest) + ': ' + str(position))
```
